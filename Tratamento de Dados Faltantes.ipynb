## Tratamento de Dados Faltantes

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')

dados.head()
```

```python
# Retorna um dataframe com booleanos, True se for nulo e False se tiver dados
dados.isnull()
```

```python
# Retorna um dataframe com booleanos, False se for nulo e True se tiver dados
dados.notnull()
```

```python
# Monstra informações sobre os dados
dados.info()
```

```python
# Retorna os registros com nulo no campo 'Valor'.
dados[dados['Valor'].isnull()]
```

Como o que nos importa para estas análises são os valores dos aluguéis, vamos deletar os registros que não tem valor. 

Para isso usamos a função **.dropna()**.

```python
# Removendo registros que não tem valor.
A = dados.shape[0]
dados.dropna(subset=['Valor'], inplace=True)
B = dados.shape[0]

#Total de registros excluídos.
A - B
```

```python
dados[dados['Valor'].isnull()]
```

## Tratamento de Dados Faltantes (Continuação)

```python
dados[dados['Condominio'].isnull()]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())

dados[selecao]
```

Quando temos dados boleanos uma forma de inverter o valor é usando o 'tio' **~** conforme exemplo a seguir.

```python
# Removendo registros que são nulos
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]

#Total de registros excluídos.
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

Como temos ainda 1068 registros nulos mas que não podemos excluir vamos colocar um valor padrão. Neste caso será 0.

```python
dados.fillna(0, inplace=True)
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
#Outra Maneira de fazer para multiplos atributos é:
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

```python
dados.info()
```

```python
# Substituindo dados antigos pelos tratados.
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index=False)
```
